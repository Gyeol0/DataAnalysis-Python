```python
from sklearn.model_selection import train_test_split
Data=pd.read_csv('Pima Indian Diabetes.csv')
xy=np.loadtxt('Pima Indian Diabetes.csv',delimiter=',',skiprows=1,dtype=np.float32)
x_data=xy[:,0:-1]
y_data=xy[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size= 0.1, random_state=1111) ##9:1 분할
X=tf.placeholder(tf.float32,shape=[None,8])
Y=tf.placeholder(tf.float32,shape=[None,1])
W=tf.Variable(tf.random_normal([8, 1],mean=0.001, stddev=0.01))
b=tf.Variable(tf.random_normal([1]))
hypo=tf.sigmoid(tf.matmul(X,W)+b)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
predicted=tf.cast(hypo>0.5,dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))
cost=-tf.reduce_mean(Y*tf.log(hypo)+(1-Y)*tf.log(1-hypo))
```

```python
train=optimizer.minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10001): 
    sess.run(train,feed_dict={X:X_train,Y:y_train})
    if i %200==0:
        print(i, sess.run(cost,feed_dict={X:X_train,Y:y_train}))
```

```python
h,c,a=sess.run([hypo,predicted,accuracy],feed_dict={X:X_test,Y:y_test})
print('hypo: ',h,'\nCorrect : ',c,'\nAccuracy : ',a)
```
